# Activation functions


```python
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-10,10,100)

#sigmoid function
y1=1./(1+np.exp(-x))

#Tangnet Hyperbolic

y2=np.tanh(x)

#RelU
y3=np.maximum(0,x)
```

```python
#activation function

plt.style.use('ggplot')
plt.figure(figsize=(15,3))

plt.subplot(1,3,1)
plt.plot(x,y1,label='sigmoid')

plt.subplot(1,3,2)
plt.plot(x,y2,label='tanh')

plt.subplot(1,3,3)
plt.plot(x,y3,label='Relu')
plt.show();
```
