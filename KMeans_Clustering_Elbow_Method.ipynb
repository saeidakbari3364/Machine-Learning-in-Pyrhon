```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('/content/Mall_Customers.csv')
```

```python
df.head()
```

```python
x=df.iloc[:,[3,4]].values
```

```python
wcss_list=[]
for i in range(1,11):
  kms=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kms.fit(x)
  wcss_list.append(kms.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1,11),wcss_list)
plt.title('The Elbow Method Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('wcss_list')
plt.show()
```

```python
kms1=KMeans(n_clusters=5,init='k-means++',random_state=42)
y_predict=kms1.fit_predict(x)
```

```python
plt.scatter(x[y_predict == 0, 0], x[y_predict == 0, 1], s=100, c='blue',label='Cluster 1')
plt.scatter(x[y_predict == 1, 0], x[y_predict == 1, 1], s=100, c='green',label='Cluster 2')
plt.scatter(x[y_predict == 2, 0], x[y_predict == 2, 1], s=100, c='red',label='Cluster 3')
plt.scatter(x[y_predict == 3, 0], x[y_predict == 3, 1], s=100, c='cyan',label='Cluster 4')
plt.scatter(x[y_predict == 4, 0], x[y_predict == 4, 1], s=100, c='magenta',label='Cluster 5')
plt.scatter(kms1.cluster_centers_[:, 0], kms1.cluster_centers_[:, 1],s=100, c='yellow', label='Centroid')
plt.title('Clusters of customers')
plt.xlabel('Annual income (k$)')
plt.ylabel('Spending score (1-100)')
plt.legend()
plt.show()
```
