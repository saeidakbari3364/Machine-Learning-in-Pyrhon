```python
import pandas as pd
```

```python
df=pd.DataFrame({'column_one':['1.1','2.2','3.3'],'column_two':['2.2','3.3','4.4'],'column_three':['7.7','8.8','-']})
df.head()
```

```python
df.dtypes
```

```python
df.astype({'column_one':'float','column_two':'float'}).dtypes
```

```python
pd.to_numeric(df.column_three,errors='coerce')
```

```python
pd.to_numeric(df.column_three,errors='coerce').fillna(0)
```

```python
#all columns of dataframe
df=df.apply(pd.to_numeric,errors='coerce').fillna(0)
```
