```python
import matplotlib.pyplot as plt
```

```python
#import numpy library for get datat to our plots
```

```python
import numpy as np
```

```python
#creating data
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
#Plotting the data
```

```python
#Sine Wave
```

```python
plt.plot(x,y)
plt.show()
```

```python
#Cosine Wave
```

```python
plt.plot(x,z)
plt.show()
```

```python
#adding title,x-axis,yaxis labels
```

```python
plt.plot(x,y)
plt.title("Sin Plot")
plt.xlabel("angle")
plt.ylabel('sin value')
plt.show()
```

```python
#Parabola
```

```python
x=np.linspace(-10,10,20)
y=x**2
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

```python
#Bar Plot
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
Languages=['English','French','Spanish','German']
people=[100,120,50,40]
ax.bar(Languages,people)
plt.xlabel('Languages')
plt.ylabel('Pepole')
plt.show()
```

```python
#Pie chart
```

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
Languages=['English','French','Spanish','German']
people=[100,120,50,40]
ax.pie(people,labels=Languages,autopct='%1.1f%%')
plt.show()
```

```python
#Scatter Plot
```

```python
x=np.linspace(0,10,50)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
