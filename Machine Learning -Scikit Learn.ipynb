```python
fig = plt.figure(figsize=(8, 8))  # figure size in inches
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

# plot the digits: each image is 8x8 pixels
for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
    ax.imshow(digits.images[i], cmap=plt.cm.binary, interpolation='nearest')
    
    # label the image with the target value
    ax.text(0, 7, str(digits.target[i]), bbox=dict(facecolor='red', alpha=0.1))
```

```python
from sklearn.datasets import fetch_olivetti_faces
```

```python
faces=fetch_olivetti_faces()
fig=plt.figure(figsize=(10,10))
fig.subplots_adjust(left=0,right=1,bottom=0,top=1,hspace=0.05,wspace=0.05)
for i in range(100):
        ax=fig.add_subplot(10,10,i+1,xticks=[],yticks=[])
        ax.imshow(faces.images[i],cmap=plt.cm.bone,interpolation='nearest')
```

# Classifying Faces

```python
X,y=faces.data,faces.target
```

```python
from sklearn.neural_network import MLPClassifier
clf=MLPClassifier()
```

```python
clf.fit(X,y)
```

```python
y_pred=clf.predict(X)
```

```python
accuracy=np.mean(y_pred==y)
print("accuracy=%.2f" %accuracy)
```

# Split Train and Test data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(faces.data,faces.target,test_size=0.25)
```

```python
print('Number of Train Data:%d' %X_train.shape[0])
print('Number of Test Data:%d' %X_test.shape[0])
```

```python
clf=MLPClassifier(hidden_layer_sizes=(100,100),alpha=0.002,verbose=True,max_iter=200)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
```

```python
print("accuracy =%.2f"%np.mean(y_pred==y_test))
```
