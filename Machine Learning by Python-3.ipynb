# Ploting

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.arange(0,3*np.pi,0.1)
```

```python
print(x)
```

```python
y1=np.sin(x)
plt.plot(x,y1)
```

```python
y2=np.cos(x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin and cos')
plt.legend(['sin','cos'])
```

```python
plt.plot(x,y1,'r--',x,y2,'k:')
plt.legend(['Sin','Cos'])
```

```python
x=np.arange(0,3*np.pi,0.1)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
y4=np.abs(x)
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title('sin')
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('cos')
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('tan')
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('abs')
plt.show()
```

```python
x=np.linspace(-1.4,1.4,30)
plt.subplot(2,2,1)
plt.plot(x,x)
plt.title('y=x')
plt.subplot(2,2,2)
plt.plot(x,x**2)
plt.title('y=X^2')
plt.subplot(2,2,3)
plt.plot(x,x**3)
plt.title('y=x^3')
plt.subplot(2,2,4)
plt.plot(x,x**4)
plt.title('y=x^4')
plt.show()
```

# Saving a figure

```python
x=np.linspace(-2,2,500)
y=x**2
plt.plot(x,y,'r--')
plt.xlabel('$X$')
plt.ylabel('$y=x^2$')
plt.title('Squre Function')
#save the figure
plt.savefig('x2.png')
```
